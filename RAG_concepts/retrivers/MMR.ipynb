## MMR - maximal marginal relevance
its designed to reduce redundacy in the retrieved results while maintaining high relevance to the query

```python
from langchain_ollama import OllamaEmbeddings
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma, FAISS
```

```python
docs= [
    Document(
        page_content='Langchain helps developers build LLM application'
    ),
    Document(
        page_content='Chroma is a vector store optimised for LLM-based search'
    ),
    Document(
        page_content='Embeddings converts text into high dimensional vectors'
    ),
]
```

```python
embed_model = OllamaEmbeddings(model = 'nomic-embed-text')

vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embed_model
)
```

```python
retriever = vector_store.as_retriever(
    search_type = 'mmr',      # this enables MMR
    search_kwargs = {"k":3,"lamba_mult" : 0.5}    #k=top results, lambda_mul = relevanve diversity balance
                                                                            # 0 for most diversion
)
```

```python
res = retriever.invoke("what helps to build LLM apps")
for i in res:
    print(i.page_content)
```
